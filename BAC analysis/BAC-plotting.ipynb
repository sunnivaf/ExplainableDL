```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Importing data
df = pd.read_csv('BACs.csv')
print(df)
df = df.iloc[:-1,:] #Remove M/F row in plots
df.dropna() #Drop NaNs in plots
print(df)
```

```python
# Creating male/female dataframes
df_males = df.iloc[:,[0,6,8,9,11,12,14,17,18,19,20]].copy()
print(df_males)
df_females = df.iloc[:, [0,1,2,3,4,5,7,10,13,15,16]].copy()
print(df_females)
```

```python
# Finding and adding means for each BAC measurement
df['mean'] = df.iloc[:, 1:].mean(axis=1)
print(df)
df_males['mean'] = df_males.iloc[:, 1:].mean(axis=1)
print(df_males)
df_females['mean'] = df_females.iloc[:, 1:].mean(axis=1)
print(df_females)
```

```python
lines = df.plot.line(x='Time', y='mean',title='BAC curves for all participants', label='All', ylabel=u'BACs [‰]', style='o-')#.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
lines2 = df_males.plot.line(ax=lines, y='mean',x='Time', label='Males',style='o-')
lines3 = df_females.plot.line(ax=lines2, y='mean', x='Time', label='Females', xlabel='Time after alcohol ingestion [min]',style='o-').get_figure().savefig('BACs-mean.png')
```

```python
#BAC curves with start value
df_copy = df.copy()
df_m_copy = df_males.copy()
df_f_copy = df_females.copy()

new_row = pd.DataFrame({'Time':-24, 'P1':0.0, 'P2':0.0,'P3':0, 'P4':0.0,'P5':0.0,'P6':0.0,'P7':0.0,'P8':0.0,'P9':0.0,'P10':0.0,'P11':0.0,'P12':0.0,'P13':0.0,'P14':0.0,'P15':0.0,'P16':0.0,'P17':0.0,'P18':0.0,'P19':0.0,'P20':0.0, 'mean':0.0},index =[0])
new_row_m = pd.DataFrame({'Time':-24,'P6':0.0,'P8':0.0,'P9':0.0,'P11':0.0,'P12':0.0,'P14':0.0,'P17':0.0,'P18':0.0,'P19':0.0,'P20':0.0, 'mean':0.0},index =[0])
new_row_f = pd.DataFrame({'Time':-24, 'P1':0.0, 'P2':0.0,'P3':0, 'P4':0.0,'P5':0.0,'P7':0.0,'P10':0.0,'P13':0.0,'P15':0.0,'P16':0.0, 'mean':0.0},index =[0])

# simply concatenate both dataframes
df_copy = pd.concat([new_row, df_copy]).reset_index(drop = True)
df_m_copy = pd.concat([new_row_m, df_m_copy]).reset_index(drop = True)
df_f_copy = pd.concat([new_row_f, df_f_copy]).reset_index(drop = True)
df_f_copy.head()
```

```python
lines4 = df_copy.plot.line(x='Time', y='mean',title='BAC curves for all participants', label='All', ylabel=u'BACs [‰]', style='o-')#.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
lines5 = df_m_copy.plot.line(ax=lines4, y='mean',x='Time', label='Males',style='o-')
lines6 = df_f_copy.plot.line(ax=lines5, y='mean', x='Time',style='o-', label='Females', xlabel='Time after alcohol ingestion [min]').get_figure().savefig('BACs-mean-from-zero.png')
```
