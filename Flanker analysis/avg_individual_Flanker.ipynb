```python
import pandas as pd
import numpy as np
```

```python
def round_list(list, decimal):
    new_list = []
    for number in list:
        new_list.append(round(number, decimal))
    return new_list
```

```python
#Get data from participants
placebo = pd.read_csv('../Group level/Placebo.csv')
alcohol = pd.read_csv('../Group level/Alcohol.csv')

# Fix types and change None to max RT [ms]
alcohol['RT'] = alcohol['RT'].replace('None',2000).astype(int)
placebo['RT'] = placebo['RT'].replace('None',2000).astype(int)

alcohol = alcohol.astype({'trialtype':'string'})
placebo = placebo.astype({'trialtype':'string'})
```

```python
# Get all participant data
def extract_data(participant):
    p = placebo.loc[placebo['subject_id'] == participant]
    a = alcohol.loc[alcohol['subject_id'] == participant]
    return (p, a)
```

```python
# Find RTs and accs + SDs
def find_means(placebo_df, alcohol_df):
    #RTs
    means_rt = [placebo_df['RT'].mean(), alcohol_df['RT'].mean()]
    stds_rt = [np.std(placebo_df['RT']), np.std(alcohol_df['RT'])]
    rt = [means_rt, stds_rt]

    #Accuracies
    means_acc = [placebo_df['accuracy'].mean(), alcohol_df['accuracy'].mean()]
    stds_acc = [np.std(placebo_df['accuracy']), np.std(alcohol_df['accuracy'])]
    accuracy = [means_acc, stds_acc]

    return (rt, accuracy)
```

```python
def get_values():
    columns = ['participant', 'p_RT', 'p_RT std', 'p_accuracy', 'p_accuracy std', 'a_RT', 'a_RT std','a_accuracy','a_accuracy std']
    data = []
    for x in range(1,21):
        placebo_vals, alcohol_vals = extract_data(x)
        rts, accs = find_means(placebo_vals,alcohol_vals)
        participant_data = [x, rts[0][0],rts[1][0],accs[0][0],accs[1][0],rts[0][1],rts[1][1],accs[0][1],accs[1][1]]
        data.append(participant_data)
    return pd.DataFrame(data=data,columns=columns)
```

```python
averages = get_values()
averages
```
