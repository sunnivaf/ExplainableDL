```python
import pandas as pd
import glob
```

```python
## This file (combine-CSVs.ipynb) must be in the same directory as the CSV-files ##
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
df_append = pd.DataFrame()

#append all files together
for file in csv_files:
            df_temp = pd.read_csv(file) # add 'header=1' if header is on second line
            df_append = df_append.append(df_temp, ignore_index=True)
df_append
```

```python
df_append.to_csv('Baseline-All_CSVs.csv',index=False)
```
